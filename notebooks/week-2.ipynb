```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
ktm_img = cv2.imread('../assets/ktm-farrel.jpg')

if ktm_img is None:
    raise Exception('Image is not found!')

ktm_img_rgb = cv2.cvtColor(ktm_img, cv2.COLOR_BGR2RGB)
plt.imshow(ktm_img_rgb)
```

```python
# check image shape
ktm_img_rgb.shape
```

```python
# let's target the barcode on that image
# based on cartesius diagram, the barcode start around (1700, 700) until (1750, 950)

ktm_img_rgb_copy_1 = ktm_img_rgb.copy()

start_coordinate = [1700, 700] # (x, y)
end_coordinate = [1964, 950] # (x, y)

ktm_img_rgb_copy_1[
    start_coordinate[1]:end_coordinate[1], 
    start_coordinate[0]:end_coordinate[0]
] = [123, 45, 210] # replace with purple color

plt.imshow(ktm_img_rgb_copy_1)
```
